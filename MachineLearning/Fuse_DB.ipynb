Assuming you have parametrised multiple data bases, you might want to fuse them into a single now that they are much lighter. You can do so by using this notebook. Be aware that a final data frame that is too big might cause some problems !

```python
import numpy as np
import pandas as pd
import os
```

```python
sample = 'Iron/'
curve = 'Half/'
PISN_split = '-2'
model = '_poly'
purpose = 'test'
other = '4p/' #"BigSimulation/"


path_files = "../../plasticc/param_table/"+sample+curve+other
path_save = "../../plasticc/param_table/"+sample+curve+other+purpose+PISN_split+model+"FUSED.pkl"
```

```python
all_filenames = np.array(os.listdir(path_files))
ispkl=[purpose+PISN_split+model in f for f in all_filenames]
all_filenames
fused = pd.concat([pd.read_pickle(path_files + f) for f in all_filenames[ispkl]])

fused.to_pickle(path_save)
```
